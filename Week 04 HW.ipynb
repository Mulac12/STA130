```python
#4
import numpy as np

sample = np.array([12, 7, 22, 13, 16, 9, 21, 19, 8, 10])

def bootstrap_confidence_interval(sample, statistic_func=np.mean, num_bootstrap_samples=1000, confidence_level=0.95):
    """
    Function to calculate a bootstrap confidence interval for a specified statistic (e.g., mean or median).
    
    
    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    
    bootstrap_statistics = []
    
    for i in range(num_bootstrap_samples):
        bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)
        
        stat_value = statistic_func(bootstrap_sample)
        bootstrap_statistics.append(stat_value)
    
    lower_percentile = (1 - confidence_level) / 2 * 100
    upper_percentile = (1 + confidence_level) / 2 * 100
    
    lower_bound = np.percentile(bootstrap_statistics, lower_percentile)
    upper_bound = np.percentile(bootstrap_statistics, upper_percentile)
    
    return lower_bound, upper_bound

ci_mean = bootstrap_confidence_interval(sample, statistic_func=np.mean)
print(f"95% Bootstrap Confidence Interval for the Mean: {ci_mean}")
```

```python
#5: We need to distinguish between the population parameter and the sample statistic because confidence intervals are used to estimate the population parameter (which is typically unknown) based on the sample statistic (which is calculated from the sample). The sample statistic is subject to variability, as it can change depending on the sample taken, while the population parameter is a fixed value that we're trying to infer. Confidence intervals account for this variability and provide a range within which the population parameter is likely to fall, based on the sample data.
```
