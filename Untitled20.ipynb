```python
#3. 
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import seaborn as sns
df = sns.load_dataset('titanic')
df = df.dropna(subset=['survived', 'age', 'sex'])
df['sex_binary'] = df['sex'].map({'male': 0, 'female': 1})
additive_formula = 'survived ~ age + sex_binary'
additive_model = smf.logit(additive_formula, data=df).fit()
print(additive_model.summary())
interaction_formula = 'survived ~ age * sex_binary'
interaction_model = smf.logit(interaction_formula, data=df).fit()
print(interaction_model.summary())
# Simulate data for plotting
import numpy as np

# Create a range of ages
age_range = np.linspace(df['age'].min(), df['age'].max(), 100)

# Predict survival probabilities for males and females
male_preds = additive_model.predict(pd.DataFrame({'age': age_range, 'sex_binary': 0}))
female_preds = additive_model.predict(pd.DataFrame({'age': age_range, 'sex_binary': 1}))

# Create a DataFrame for plotting
plot_df = pd.DataFrame({
    'age': np.concatenate([age_range, age_range]),
    'survival_probability': np.concatenate([male_preds, female_preds]),
    'sex': ['Male']*100 + ['Female']*100
})

# Plot the data
fig = px.scatter(df, x='age', y='survived', color='sex',
                 labels={'survived': 'Survival', 'age': 'Age', 'sex': 'Sex'},
                 title='Additive Model: Survival by Age and Sex')

# Add lines
fig.add_traces(px.line(plot_df, x='age', y='survival_probability', color='sex').data)

fig.show()
# Predict survival probabilities with interaction
male_preds_int = interaction_model.predict(pd.DataFrame({'age': age_range, 'sex_binary': 0}))
female_preds_int = interaction_model.predict(pd.DataFrame({'age': age_range, 'sex_binary': 1}))

# Create a DataFrame for plotting
plot_df_int = pd.DataFrame({
    'age': np.concatenate([age_range, age_range]),
    'survival_probability': np.concatenate([male_preds_int, female_preds_int]),
    'sex': ['Male']*100 + ['Female']*100
})

# Plot the data
fig_int = px.scatter(df, x='age', y='survived', color='sex',
                     labels={'survived': 'Survival', 'age': 'Age', 'sex': 'Sex'},
                     title='Interaction Model: Survival by Age and Sex with Interaction')

# Add lines
fig_int.add_traces(px.line(plot_df_int, x='age', y='survival_probability', color='sex').data)

fig_int.show()
```

5. Model Complexity vs. Generalization:

Model 3 provides a simpler model with fewer parameters, which may generalize better to new data.
Model 4 is more complex, capturing intricate relationships in the training data but potentially overfitting.
Overfitting:

Occurs when a model learns the noise in the training data rather than the underlying pattern.
Leads to poor performance on new, unseen data.
Evaluating Models:

In-Sample R-squared: Measures fit on the training data.
Out-of-Sample R-squared: Measures predictive performance on new data.
A significant discrepancy between the two can indicate overfitting.
Practical Implications:

Adding too many predictors and interaction terms can lead to models that are computationally intensive and less interpretable.
It's crucial to balance model complexity with the need for generalization.

By progressively refining the model specifications and carefully selecting which variables and interactions to include, we aim to build a model that provides accurate predictions of HP while avoiding the pitfalls of overfitting and multicollinearity. Each model builds upon the previous one, incorporating new insights and adjustments based on theoretical considerations and empirical evidence.

```python
#8
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
pokeaman = pd.read_csv(url)

# Replace NaN in 'Type 2' with 'None'
pokeaman['Type 2'].fillna('None', inplace=True)

# Convert 'Legendary' to binary (if not already)
pokeaman['Legendary'] = pokeaman['Legendary'].astype(int)
model3_formula = 'HP ~ Attack + Defense'
model4_formula = (
    'HP ~ Attack * Defense * Speed * Legendary * Q("Sp. Def") * Q("Sp. Atk")'
)
# Initialize lists to collect R-squared values
model3_in_sample_r2 = []
model3_out_sample_r2 = []
model4_in_sample_r2 = []
model4_out_sample_r2 = []

# Number of iterations
n_iterations = 100

for i in range(n_iterations):
    # Randomly split the data without setting a seed
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # --- Model 3 ---
    # Fit the model on training data
    model3_spec = smf.ols(formula=model3_formula, data=pokeaman_train)
    model3_fit = model3_spec.fit()
    
    # In-sample R-squared
    model3_in_sample_r2.append(model3_fit.rsquared)
    
    # Out-of-sample R-squared
    y_true = pokeaman_test['HP']
    y_pred = model3_fit.predict(pokeaman_test)
    r = np.corrcoef(y_true, y_pred)[0, 1]
    model3_out_sample_r2.append(r ** 2)
    
    # --- Model 4 ---
    # Fit the model on training data
    model4_spec = smf.ols(formula=model4_formula, data=pokeaman_train)
    model4_fit = model4_spec.fit()
    
    # In-sample R-squared
    model4_in_sample_r2.append(model4_fit.rsquared)
    
    # Out-of-sample R-squared
    y_pred = model4_fit.predict(pokeaman_test)
    r = np.corrcoef(y_true, y_pred)[0, 1]
    model4_out_sample_r2.append(r ** 2)
# Prepare data for plotting
data = {
    'Model 3 In-Sample R^2': model3_in_sample_r2,
    'Model 3 Out-of-Sample R^2': model3_out_sample_r2,
    'Model 4 In-Sample R^2': model4_in_sample_r2,
    'Model 4 Out-of-Sample R^2': model4_out_sample_r2
}

df_results = pd.DataFrame(data)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot for Model 3
axes[0].boxplot([df_results['Model 3 In-Sample R^2'], df_results['Model 3 Out-of-Sample R^2']],
                labels=['In-Sample R^2', 'Out-of-Sample R^2'])
axes[0].set_title('Model 3 Performance')
axes[0].set_ylim(0, 1)
axes[0].grid(True)

# Boxplot for Model 4
axes[1].boxplot([df_results['Model 4 In-Sample R^2'], df_results['Model 4 Out-of-Sample R^2']],
                labels=['In-Sample R^2', 'Out-of-Sample R^2'])
axes[1].set_title('Model 4 Performance')
axes[1].set_ylim(0, 1)
axes[1].grid(True)

plt.show()
```
